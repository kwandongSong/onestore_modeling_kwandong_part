```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


data = pd.read_csv('data_sample.csv')

#------------기초 통계량 구하기 -------함수화

raw_data=data.loc[:,['prod_amt','new_id']]

def get_prod_all(what,prod_all):
        if what=='sum':
            prod_all=pd.pivot_table(raw_data,index=["new_id"],aggfunc=np.sum)
            prod_all.rename(columns={'prod_amt':'prod_sum_dt'},inplace=True)
        elif what=='mean':
            prod_all['prod_mean_dt']=pd.pivot_table(raw_data,index=["new_id"],aggfunc=np.mean)
        elif what=='max':
            prod_all['prod_max_dt']=pd.pivot_table(raw_data,index=["new_id"],aggfunc=np.max)
        elif what=='min':
            prod_all['prod_min_dt']=pd.pivot_table(raw_data,index=["new_id"],aggfunc=np.min)
        elif what=='std':
            prod_all['prod_std_dt']=pd.pivot_table(raw_data,index=["new_id"],aggfunc=np.std)
        elif what=='var':
            prod_all['prod_var_dt']=pd.pivot_table(raw_data,index=["new_id"],aggfunc=np.var)
        else :
            print("----------------------Input error--------------------")
        return prod_all
    
prod_all=get_prod_all('sum',raw_data) #한번 돌려서 table 만들고

method_arr=['mean','max','min','std','var'] #변수 array
            
for i in method_arr: 
    prod_all=get_prod_all(i,prod_all)

prod_all
#data[data['prod_amt']==14400]
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
pd.set_option('display.max_colwidth', -1)
column_list=['prod_amt','prchs_tm_clsf_nm','sett_target_cpn_amt','cust_payment_amt']
data = pd.read_csv('data_190814.csv',usecols=['prod_amt','prchs_tm_clsf_nm','sett_target_cpn_amt','cust_payment_amt','new_id'])
all= data.loc[:,['new_id']]

def get_column_statistics_data(what,column_statistics_data):
        if what=='sum':
            column_statistics_data=pd.pivot_table(raw_data,index=["new_id"],aggfunc=np.sum)
            column_statistics_data.rename(columns={select_column:select_column+'_sum'},inplace=True)
            if select_column=='cust_payment_amt':
                column_statistics_data.rename(columns={'cust_payment_amt_sum': 'Monetary'},inplace=True)
        elif what=='mean':
            column_statistics_data[select_column+'_mean']=pd.pivot_table(raw_data,index=["new_id"],aggfunc=np.mean)
            if select_column=='cust_payment_amt':
                column_statistics_data.rename(columns={'cust_payment_amt_mean': 'avg_prchs_amt'},inplace=True)
        elif what=='max':
            column_statistics_data[select_column+'_max']=pd.pivot_table(raw_data,index=["new_id"],aggfunc=np.max)
        elif what=='min':
            column_statistics_data[select_column+'_min']=pd.pivot_table(raw_data,index=["new_id"],aggfunc=np.min)
        elif what=='std':
            column_statistics_data[select_column+'_std']=pd.pivot_table(raw_data,index=["new_id"],aggfunc=np.std)
        elif what=='var':
            column_statistics_data[select_column+'_var']=pd.pivot_table(raw_data,index=["new_id"],aggfunc=np.var)
        else :
            print("----------------------Input error--------------------")
        return column_statistics_data
for i in column_list:
    
    #모든 변수에 쓸 수 있게 해보자
    #prchs_tm_clsf_nm, sett_target_cpn_amt, cust_payment_amt

    #select_column='prod_amt'
    #select_column='prchs_tm_clsf_nm'
    #select_column='sett_target_cpn_amt'
    #select_column='cust_payment_amt'
    select_column=i

    raw_data = data.loc[:,[select_column,'new_id']]

    column_statistics_data=get_column_statistics_data('sum',raw_data) #한번 돌려서 table 만들고

    method_arr=['mean','max','min','std','var'] #변수 array

    for i in method_arr: 
        column_statistics_data=get_column_statistics_data(i,column_statistics_data)

    all=pd.merge(all,column_statistics_data,on='new_id')

all=all.drop_duplicates()
all

all.to_csv('data_kwandong.csv',index_label=False)
```
